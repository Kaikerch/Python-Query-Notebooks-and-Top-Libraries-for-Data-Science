```python
for i in idades:
  permissao_para_dirigir(i)
```

```python
def permissao_para_dirigir(idades):
  
  for idade in idades:

    if (idade < 18):
      print(f"{idade} anos não tem permissão para dirigir.")
    else:
      print(f"{idade} pode dirigir!")
```

```python
permissao_para_dirigir(idades)
```

## **Booleano**

```python
idade = 27
idade >= 18
```

```python
idade = 15
idade >= 18
```

```python
permissoes = []
idades = [18, 56, 70, 10]
```

```python
def verifica_se_pode_dirigir(idades, permissoes):
  
  for idade in idades:

    if (idade >= 18):
      permissoes.append(True)
    else:
      permissoes.append(False)
```

```python
verifica_se_pode_dirigir(idades, permissoes)
```

```python
permissoes
```

```python
for permissao in permissoes:

  if (permissao == True):
    print('Pode dirigir!')
  else:
    print('Não pode dirigir') 
```
